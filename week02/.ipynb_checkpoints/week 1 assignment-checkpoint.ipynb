# Community Redevelopment Agency Project Boundary
For this assignment, I will be looking at the data for the community redevelopment agency project boundaries provided by the Los Angeles City GeoHub and where the most redevelopment is happening. This dataset defines the different neighborhoods that are designated to be redeveloped, the number of acres, and the regions in which they are happening. The regions will allow me to understand which regions are experiencing the most redevelopment. 

```python
#importing the different packages needed to perform these functions
import pandas as pd
import geopandas as gpd
```

```python
#importing my dataset
df = pd.read_csv("https://raw.githubusercontent.com/leedayounglee/up221-dayoung/main/W1.2/Community_Redevelopment_Agency_Project_Boundary.csv")
```

```python
#defining the "Community_Redevelopment_Agency_Project" file as crap for short and easy navigation.
crap = df
crap
```

## **Head**
Performing the .head function outputs the first five values in a dataset. This is useful for confirming that the imported data has successfully imported the correct number of columns.

```python
crap.head()
```

## **Shape**
The .shape function allows us to see the dimensions of our data set in terms of (columns, rows). 

```python
crap.shape
```

## **Trim**
I've trimmed the data to  show only the data that is relevant (the neighborhoods, region, acres). 

```python
crap_trim = crap[['REDEVNM', 'REGION', 'ACRES']]
crap_trim.head()
```

## **Info**
Before we continue with the data, we want to make sure that the 'ACRES' column is a float. If it were just an object (string), then it would not be plottable. 

```python
crap_trim.info()
```

## **Plot**
Now that I've confirmed that 'ACRES' is a float, I want to format the dataset to be in descending order, so that it will be visually easier to follow. 

```python
#descending values
crap_trim = crap_trim.sort_values(by = ['ACRES'], ascending = False)
```

```python
#plot bar
crap_trim.plot.bar(
    x = 'REDEVNM', 
    y = 'ACRES', 
    title = 'Acres of Redevelopment by Neighborhoods in Los Angeles'
)    
```

## **QUERY**
We want to see which regions are being the most gentrified. We will first try to determine it by neighborhoods in acres, and then we will approach it from a different angle by value counting the number of regions that pop up in the dataset.

```python
crap_trim['COLOR'] = ''
```

```python
crap_trim.head()
```

```python
crap_trim.loc[crap_trim['REGION'] == 'South Los Angeles', 'color'] = 'orange'
```

```python
crap_trim.loc[crap_trim['REGION'] == 'South Los Angeles']
```

```python
crap_trim.plot.bar(
    x = 'REGION', 
    y = 'ACRES', 
    title = 'Acres of Redevelopment by Neighborhoods in Los Angeles'
)    
```

```python
crap_trim
```

```python
#query. We are querying the number of acres being redeveloped, and labeling them as 'gentrify'. 
#Here, we are making the assumption that redevelopment boundaries with more than 100 acres are the areas being the most gentrified.
gentrify = crap_trim[crap_trim['ACRES']>100.000000]
gentrify
```

```python
#Another way to approach this is to see the number of regions that pop up within this dataset.
crap_trim_count = crap['REGION'].value_counts()
crap_trim_count
```

# **Conclusion**
From the query portion, we see that even though the East Valley had the highest number of acres in the list, South Los Angeles is the region that is experiencing the most redevelopment within the city of Los Angeles. These nuances are important when exploring datasets because the different ways we approach the data can lead to different results and analyses. Context is key in these situations.
